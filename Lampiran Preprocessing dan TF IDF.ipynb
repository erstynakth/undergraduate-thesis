# Text Cleansing

```python
# Text Cleansing

def cleansing(text):
    text = re.sub(r"\s+", " ", text, flags=re.UNICODE)
    text = re.sub(r'@\S*\s?|#\S*\s?', '', text) # username, mention, hashtags
    text = re.sub(r'[h][t][t][p][s][:][/][/][t][.][c][o][/][a-zA-Z0-9][^\s]{0,}',' ', text) # url
    text = re.sub(r'[0-9]', ' ', text) # angka
    text = emoji_to_words(text, lang = 'id', delimiter = (' ', ' ')) # emoji
    text = (unicodedata.normalize("NFD", text).encode("ascii", "ignore").decode("ascii"))
    text = re.sub(r'[\!\"\‚Äù\$\%\&\'\(\)\*\+\,\-\.\/\:\;\<\=\>\?\[\\\]\^\_\`\{\|\}\~\‚Äì]', ' ', text) # karakter spesial / tanda baca
    text = text.lstrip() # menghilangkan spasi di depan kalimat
    return text
```

```python
sample['cleansing'] = [cleansing(x) for x in sample['tweet'].values]
```

# Text Normalization

```python
# Text Normalization : ejaan kata, ubah kata tidak baku jadi baku

def normalization(text):
    text = replace_word_elongation(text) # mengubah kata yang belakangnya terlalu panjang
    text = replace_slang(text) # mengubah kata tidak baku jadi baku
    return text
```

# Stemming

```python
# Stemming

factory = StemmerFactory()
stemmer = factory.create_stemmer()

def stemming(text):
    text = stemmer.stem(text)
    return text
```

# Case Folding dan Tokenization

```python
# Case Folding dan Tokenization

def foldtoken(text):
    text = text.lower() # Case folding (huruf kecil semua)
    text = re.split(r'\s+', text.rstrip()) # Tokenization
    return text
```

```python
sample['tokenized'] = [foldtoken(x) for x in sample['stemmed'].values] 
```

```python
df['tokenized'] = [foldtoken(x) for x in df['stemmed'].values] 
```

# Filtering

```python
# Create Stopwords Dictionary
from indoNLP.preprocessing import STOPWORDS
STOPWORDS.update(StopWordRemoverFactory().get_stop_words())
added_stopwords = ['dips', 'jends', 'ugm fes', 'upnjog', 'nes', 'cuma', 'cm', 'braw', 'nya', 'pem']
added_stopwords = set(added_stopwords)

STOPWORDS.update(added_stopwords)

def filtering(text):
    return [text for text in text if text not in STOPWORDS]
```

```python
sample['filtered'] = sample['tokenized'].apply(filtering)
```

```python
df['filtered'] = df['tokenized'].apply(filtering)
```

```python
def concate(text):
  text = " ".join([x for x in text if x not in string.punctuation])
  return text
```

```python
sample['preprocessed'] = sample['filtered'].apply(lambda x: concate(x))
```

```python
tf = wordcount.divide(np.sum(wordcount, axis=1), axis=0)
```
