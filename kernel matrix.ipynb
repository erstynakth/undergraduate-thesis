# Kernel Matrix for AB (+Negatif -Netral)

```python
train_AB = pd.read_excel(r'C:\Users\ASUS\skripsi\train_AB.xlsx')
```

```python
train_AB = train_AB.drop(['Unnamed: 0'], axis = 1)
train_AB
```

```python
AB = train_AB.to_numpy()
AB
```

```python
df_AB = pd.DataFrame(kernel_AB)
df_AB.to_excel(r"C:\Users\ASUS\skripsi\kernel_AB.xlsx")
```

```python
ab_1 = AB[0]
ab_2 = AB[1]
ab_3 = AB[2]
ab_959 = AB[958]

ab_11 = ab_1.dot(ab_1)
ab_12 = ab_1.dot(ab_2)
ab_13 = ab_1.dot(ab_3)
ab_959959 = ab_959.dot(ab_959)

print("Dot product x_1*x_1 adalah:", ab_11)
print("Dot product x_1*x_2 adalah:", ab_12)
print("Dot product x_1*x_3 adalah:", ab_13)
print("Dot product x_959*x_959 adalah:", ab_959959)
```
