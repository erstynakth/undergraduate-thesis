```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
```

```python
juni21_list = []

for i, tweet in enumerate(sntwitter.TwitterSearchScraper('msib since:2021-06-01 until:2021-06-30').get_items()):
    if i > 15000:
        break
    juni21_list.append([tweet.date, tweet.url, tweet.user.username, tweet.content, tweet.likeCount, tweet.retweetCount,
                        tweet.replyCount, tweet.quoteCount])
    
juni21 = pd.DataFrame(juni21_list, columns = ['date', 'link', 'username', 'tweet', 'like', 'retweet', 'reply', 'qrt'])

juni21
```

```python
juni21.to_csv("F:\Documents ASUS\MATH\#8\S.MAT 2023\Data Fix\juni21.csv")
```
